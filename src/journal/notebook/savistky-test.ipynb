```python
def ctov(c_lst):
    vlst = []
    path = '../src/Data/SMCi/circular_traj/prueba_savitsky.csv'
    df = pd.read_csv(path)
    
    for c in c_lst:
        vlst.append(df[c].values)
        
    return np.array(vlst).T
```
